```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import bar_chart_race as bcr
import warnings
warnings.filterwarnings('ignore')
import datetime as dt 
%matplotlib inline
sns.set_theme()
pd.set_option('display.max_columns', None)
import math
```

import dataset

```python
years = [str(year) for year in range(2000, 2022)]
gdp = pd.read_csv("/content/GDP.csv", usecols=['Country Name'] + years)
gdp
```

```python
gdp.set_index("Country Name", inplace=True)
```

```python
gdp = gdp.T
gdp.head()
```

```python
gdp.columns
```

```python
gdp.describe()
```

```python
gdp.info()
```

Removing null 

```python
round(gdp.isnull().sum().sort_values(ascending=False)/len(gdp)*100,2)
```

```python
missing_values=gdp.isnull().sum()
```

```python
missing_values
```

```python
columns_with_missing_values = np.where(missing_values > 0)[0]
```

```python
df = gdp.drop(gdp.columns[columns_with_missing_values], axis=1)
```

```python
df.isnull().sum()
```

```python
bcr.bar_chart_race(
    df=df,
    orientation='h',
    sort='desc',
    n_bars=6,
    fixed_order=False,
    fixed_max=True,
    steps_per_period=10,
    interpolate_period=False,
    label_bars=True,
    bar_size=.95,
    period_length=500,
    figsize=(10, 5),
    dpi=144,
    cmap='dark12',
    title='Gross Domestic Product Growth Rate',
    title_size='',
    bar_label_size=7,
    tick_label_size=7,
    scale='linear',
    writer=None,
    fig=None,
    bar_kwargs={'alpha': .7},
    filter_column_colors=False)
```
